```python
import pandas as pd
import glob, os, re

# Ruta a la carpeta con los archivos Excel
#folder_path = r'C:\ruta\a\tu\carpeta'
folder_path = "C:/Users/diego.dinatale/OneDrive - Aenza/LOCAL_MACHINE/DATA_ANALYTICS/BD VIVA/DATA DEL MERCADO CON SCRAPPING/CERCADO especial"


# Listamos todos los .xlsx en la carpeta
excel_files = glob.glob(os.path.join(folder_path, '*.xlsx'))

dfs = []
pattern = re.compile(r'^Lima Proyecto (.+?)_\d{2}_\d{2}_\d{4}')  # captura lo que haya entre "Lima Proyecto " y "_DD_MM_AAAA"

for fpath in excel_files:
    # 1) Leemos el Excel
    df = pd.read_excel(fpath)
    # 2) Eliminamos columnas vacías
    df = df.dropna(axis=1, how='all')
    # 3) Si no quedó nada, saltamos
    if df.empty or df.shape[1] == 0:
        continue

    # 4) Extraemos el nombre del proyecto del nombre de archivo
    fn = os.path.basename(fpath)
    m = pattern.match(fn)
    proyecto = m.group(1) if m else fn.replace('.xlsx','')

    # 5) Añadimos la columna con el nombre del proyecto
    df['Proyecto'] = proyecto

    dfs.append(df)

# Concatenamos todos los DataFrames en uno solo
df_concat = pd.concat(dfs, ignore_index=True)

# (Opcional) Guardamos el resultado en un nuevo Excel
out_path = os.path.join(folder_path, 'BD Lima 1-3 dorm.xlsx')
df_concat.to_excel(out_path, index=False)
df_concat.sample(5)
```
